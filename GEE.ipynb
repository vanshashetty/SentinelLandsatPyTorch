```python
import ee

# Define a function to mask clouds from Sentinel-2 imagery.
def maskS2clouds(image):
    # Select the QA60 band (Cloud probability and Cirrus). Bit 10 is opaque clouds, Bit 11 is cirrus clouds.
    qa = image.select('QA60')

    # Bits 10 and 11 are clouds and cirrus, respectively.
    cloudBitMask = 1 << 10
    cirrusBitMask = 1 << 11

    # Both flags should be set to zero, indicating clear conditions.
    mask = qa.bitwiseAnd(cloudBitMask).eq(0).And(
             qa.bitwiseAnd(cirrusBitMask).eq(0))

    # Additionally, use the SCL (Scene Classification Layer) band for more detailed cloud/shadow masking.
    # SCL values:
    # 3 = Cloud shadows
    # 8 = Medium probability clouds
    # 9 = High probability clouds
    # 10 = Cirrus
    scl = image.select('SCL')
    cloud_shadow_mask = scl.neq(3) # Not cloud shadow
    medium_cloud_mask = scl.neq(8) # Not medium probability cloud
    high_cloud_mask = scl.neq(9) # Not high probability cloud
    cirrus_mask = scl.neq(10) # Not cirrus

    # Combine all masks
    combined_mask = mask.And(cloud_shadow_mask).And(medium_cloud_mask).And(high_cloud_mask).And(cirrus_mask)

    # Return the masked image, scaled to surface reflectance values.
    return image.updateMask(combined_mask).divide(10000)

# Load Sentinel-2 surface reflectance data.
# 'COPERNICUS/S2_SR_HARMONIZED' provides Level-2A data (surface reflectance).
collection = ee.ImageCollection('COPERNICUS/S2_SR_HARMONIZED')\
    .filterDate(start_date, end_date)\
    .filterBounds(aoi)

# Filter out images with high cloud cover.
# CLOUDY_PIXEL_PERCENTAGE is a metadata property.
filtered_collection = collection.filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 20))

# Apply the cloud masking function to the filtered collection.
# The map function applies the function to each image in the collection.
masked_collection = filtered_collection.map(maskS2clouds)

# Reduce the image collection to a single image using the median reducer.
# This helps to get a composite image with less noise and remaining cloud cover.
# We also explicitly select the bands needed for NDWI calculation and visualization.
preprocessed_image = masked_collection.median().select(['B2', 'B3', 'B4', 'B8', 'B11', 'B12'])

# Print information about the preprocessed image to verify.
print("Preprocessed Sentinel-2 image information:")
print(preprocessed_image.getInfo())
```

### Explanation of Normalized Difference Turbidity Index (NDTI)

To calculate the turbidity index, we used a common formulation known as the **Normalized Difference Turbidity Index (NDTI)**. This index leverages the spectral properties of water and suspended particles in different wavelengths:

*   **Red Band (B4)**: Suspended sediments and turbid waters generally have higher reflectance in the red portion of the spectrum compared to clear water.
*   **Blue Band (B2)**: Clear water absorbs less in the blue region, but suspended sediments can increase scattering. Using the difference and sum with the red band helps to normalize for illumination and enhance the signal from turbidity.

The formula used is:

\[ NDTI = \frac{B4 - B2}{B4 + B2} \]

Where:
*   `B4` is the **Red** band from Sentinel-2.
*   `B2` is the **Blue** band from Sentinel-2.

**Interpretation:**

*   Higher `NDTI` values generally indicate **higher turbidity** (more suspended sediments) in the water.
